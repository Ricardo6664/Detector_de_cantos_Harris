```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

imagem = cv2.imread('/content/drive/MyDrive/Colab Notebooks/Visão Computacional/imagens/xadrez.webp')
copia_img = np.copy(imagem)

copia_img = cv2.cvtColor(copia_img,cv2.COLOR_BGR2RGB)
plt.imshow(copia_img)
```

```python

cinza = cv2.cvtColor(copia_img,cv2.COLOR_RGB2GRAY)

cinza = np.float32(cinza)

# Detectando os cantos
# cv2.cornerHarris(img, blockSize, ksize, k)
    # img -> Imagem de entrada;
    # blockSize -> É o tamanho da vizinhança p/ detecção dos cantos;
    # ksize -> Parâmetro de abertura da derivada Sobel;
    # k -> Parâmetro livre do detector de Harris;
dst = cv2.cornerHarris(cinza,2,3,0.04)

dst = cv2.dilate(dst,None)

plt.imshow(dst,cmap='gray')
```

```python

retirada = 0.1 * dst.max()

cantos_img = np.copy(copia_img)

for j in range(0,dst.shape[0]):
    for i in range(0,dst.shape[1]):
        if(dst[j,i]>retirada):
            # Imagem, ponto central, raio, cor, espessura
            cv2.circle(cantos_img,(i,j),1,(0,255,0),1)
plt.imshow(cantos_img)
```
